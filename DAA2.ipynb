```python
import time

def interpolation_search(arr, target):
    low, high = 0, len(arr) - 1
    probes = 0

    while low <= high and arr[low] <= target <= arr[high]:
        probes += 1

        if low == high:
            if arr[low] == target:
                return low, probes
            return -1, probes

        pos = low + int(
            ((target - arr[low]) * (high - low))
            / (arr[high] - arr[low])
        )

        if arr[pos] == target:
            return pos, probes
        elif arr[pos] < target:
            low = pos + 1
        else:
            high = pos - 1

    return -1, probes


def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    probes = 0

    while low <= high:
        probes += 1

        mid = (low + high) // 2

        if arr[mid] == target:
            return mid, probes
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1, probes


# Roll numbers from 1 to 10000
roll_numbers = list(range(1, 10001))

target = int(input("Enter roll number to search: "))

idx_i, probes_i = interpolation_search(roll_numbers, target)
idx_b, probes_b = binary_search(roll_numbers, target)

print("\nInterpolation Search")
print("--------------------")
print("Index:", idx_i)
print("Probes:", probes_i)

print("\nBinary Search")
print("-------------")
print("Index:", idx_b)
print("Probes:", probes_b)

if probes_i < probes_b:
    print("\nInterpolation Search used fewer probes.")
elif probes_i > probes_b:
    print("\nBinary Search used fewer probes.")
else:
    print("\nBoth used the same number of probes.")
Question 2: Interpolation Search on Floating-Point Numbers (0.0–1000.0)
This version generates uniformly distributed floating-point values and measures comparisons for dataset sizes 10,000, 50,000, and 100,000.

import time
import random


def interpolation_search_float(arr, target):
    low, high = 0, len(arr) - 1
    comparisons = 0

    while low <= high and arr[low] <= target <= arr[high]:
        comparisons += 1

        if low == high:
            if abs(arr[low] - target) < 1e-9:
                return low, comparisons
            return -1, comparisons

        pos = low + int(
            ((target - arr[low]) * (high - low))
            / (arr[high] - arr[low])
        )

        if abs(arr[pos] - target) < 1e-9:
            return pos, comparisons
        elif arr[pos] < target:
            low = pos + 1
        else:
            high = pos - 1

    return -1, comparisons


def analyze_dataset(size):
    # Uniform distribution between 0.0 and 1000.0
    arr = [i * (1000.0 / (size - 1)) for i in range(size)]

    target = random.choice(arr)

    start = time.perf_counter()
    index, comparisons = interpolation_search_float(arr, target)
    elapsed = (time.perf_counter() - start) * 1000

    print(f"{size:<10} {comparisons:<15} {elapsed:.6f} ms")


print("Interpolation Search on Floating-Point Data")
print("-" * 50)
print(f"{'Size':<10} {'Comparisons':<15} {'Execution Time'}")
print("-" * 50)

for size in [10000, 50000, 100000]:
    analyze_dataset(size)
```
