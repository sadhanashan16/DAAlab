```python
import random

def naive_2d_search(text, pattern):
    rows_text = len(text)
    cols_text = len(text[0])
    rows_pat = len(pattern)
    cols_pat = len(pattern[0])

    matches = []
    comparisons = 0

    # Slide the pattern over the text matrix
    for i in range(rows_text - rows_pat + 1):
        for j in range(cols_text - cols_pat + 1):

            matched = True

            for x in range(rows_pat):
                for y in range(cols_pat):
                    comparisons += 1

                    if text[i + x][j + y] != pattern[x][y]:
                        matched = False
                        break

                if not matched:
                    break

            if matched:
                matches.append((i, j))

    return matches, comparisons


# -------- Main Program --------

# Generate a random 10x10 text matrix
text = [[random.choice(['A', 'B', 'C', 'D']) for _ in range(10)] for _ in range(10)]

# Define a 3x3 pattern
pattern = [
    ['A', 'B', 'C'],
    ['B', 'C', 'D'],
    ['C', 'D', 'A']
]

# Insert the pattern into the text at a known position (for verification)
start_row = 4
start_col = 5

for i in range(3):
    for j in range(3):
        text[start_row + i][start_col + j] = pattern[i][j]

# Display Text Matrix
print("10x10 Text Matrix:")
for row in text:
    print(" ".join(row))

# Display Pattern Matrix
print("\n3x3 Pattern Matrix:")
for row in pattern:
    print(" ".join(row))

# Search Pattern
matches, comparisons = naive_2d_search(text, pattern)

# Results
print("\nPattern Found At Positions:", matches)
print("Total Element Comparisons:", comparisons)
```
